```python
%%writefile ./project21/streamlit_app.py

import streamlit as st
import openai

# 1 ìŠ¤íŠ¸ë¦¼ë¦¿ ê¸°ë³¸ í˜ì´ì§€ ì„¤ì •
def set_page_style():
    st.set_page_config(
        page_title="âœ¨ê´‘ê³  ë¬¸êµ¬ ìƒì„±ê¸°",
        page_icon="ğŸ¨",
        layout="centered"
    )
    # 2 ì»¤ìŠ¤í…€ CSS ìŠ¤íƒ€ì¼ ì ìš©
    st.markdown("""
        <style>
        .stButton > button {
            background-color: #FF6B6B;
            color: white;
            border-radius: 10px;
        }
        .stButton > button:hover {
            background-color: #FF8787;
        }
        div[data-testid="stToolbar"] {
            display: none;
        }
        </style>
    """, unsafe_allow_html=True)

def generate_ad_copy(description, api_key):
    # 3 OpenAI API í‚¤ ì„¤ì •
    client = openai.OpenAI(api_key=api_key)
    
    # 4 GPT í”„ë¡¬í”„íŠ¸ ì„¤ì •
    prompt = f"""
    ë‹¤ìŒ ìƒí’ˆì„ MZì„¸ëŒ€ë¥¼ ìœ„í•œ íŠ¸ë Œë””í•œ ê´‘ê³  ë¬¸êµ¬ 3ê°œë¡œ ë§Œë“¤ì–´ì£¼ì„¸ìš”.
    ìƒí’ˆ ì„¤ëª…: {description}
    
    - ê° ë¬¸êµ¬ì— ì´ëª¨í‹°ì½˜ í¬í•¨
    - ìµœì‹  íŠ¸ë Œë“œ ìš©ì–´ ì‚¬ìš©
    - SNS ìŠ¤íƒ€ì¼ë¡œ ì‘ì„±
    """
    
    # 5 GPT API í˜¸ì¶œ ë° ì‘ë‹µ ìƒì„±
    response = response = client.chat.completions.create(
        model="gpt-4o",
        messages=[
            {"role": "system", "content": "ë‹¹ì‹ ì€ MZì„¸ëŒ€ì˜ ì–¸ì–´ì™€ íŠ¸ë Œë“œë¥¼ ì˜ ì´í•´í•˜ëŠ” ê´‘ê³  ì¹´í”¼ë¼ì´í„°ì…ë‹ˆë‹¤."},
            {"role": "user", "content": prompt}
        ],
        temperature=0.8
    )
    return response.choices[0].message.content.strip().split('\n')
    
def main():
    # 6 í˜ì´ì§€ ìŠ¤íƒ€ì¼ ì„¤ì •
    set_page_style()
    
    # 7 ë©”ì¸ íƒ€ì´í‹€ í‘œì‹œ
    st.title("âœ¨íŠ¸ë Œë””í•œ ê´‘ê³  ë¬¸êµ¬ ìƒì„±ê¸°")
    st.write("ìƒí’ˆ ì„¤ëª…ì„ ì…ë ¥í•˜ë©´ í™í•œ ê´‘ê³  ë¬¸êµ¬ë¥¼ ë§Œë“¤ì–´ë“œë ¤ìš”ğŸ¨")
    
    # 8 ì‚¬ìš©ì ì…ë ¥ í•„ë“œ ìƒì„±
    api_key = st.text_input("OpenAI API í‚¤", type="password")
    product_description = st.text_area(
        "ìƒí’ˆ ì„¤ëª…",
        height=100,
        placeholder="ì˜ˆì‹œ: 20ëŒ€ ì—¬ì„±ì„ ìœ„í•œ í™”ì´íŠ¸ ë¯¸ë‹ˆë©€ í¬ë¡œìŠ¤ë°±. ê°€ë²¼ìš´ ë¬´ê²Œê°ì´ íŠ¹ì§•ì´ì—ìš”."
    )
    # 9 ê´‘ê³  ë¬¸êµ¬ ìƒì„± ë²„íŠ¼ ì²˜ë¦¬
    if st.button("ê´‘ê³  ë¬¸êµ¬ ìƒì„±í•˜ê¸° "):
        if not api_key or not product_description:
            st.error("API í‚¤ì™€ ìƒí’ˆ ì„¤ëª…ì„ ëª¨ë‘ ì…ë ¥í•´ì£¼ì„¸ìš”!")
            return
    
        try:
            # ë¡œë”© ìƒíƒœ í‘œì‹œ ë° ê´‘ê³  ë¬¸êµ¬ ìƒì„±
            with st.spinner("ê´‘ê³  ë¬¸êµ¬ ìƒì„± ì¤‘..."):
                ad_copies = generate_ad_copy(product_description, api_key)
                
                # ì„±ê³µ ë©”ì‹œì§€ í‘œì‹œ
                st.success("ê´‘ê³  ë¬¸êµ¬ê°€ ìƒì„±ë˜ì—ˆìŠµë‹ˆë‹¤!ğŸ‰")
                st.markdown("---")
                
                # ê° ê´‘ê³  ë¬¸êµ¬ë¥¼ ê¹”ë”í•˜ê²Œ í‘œì‹œ
                for copy in ad_copies:
                    if copy.strip():
                        st.info(copy)
                st.caption("ë” ë§ì€ ì•„ì´ë””ì–´ê°€ í•„ìš”í•˜ë‹¤ë©´ ë²„íŠ¼ì„ ë‹¤ì‹œ ëˆŒëŸ¬ë³´ì„¸ìš”!ğŸ’¡")
                
        except Exception as e:
            # ì—ëŸ¬ ë©”ì‹œì§€ í‘œì‹œ
            st.error("ì˜¤ë¥˜ê°€ ë°œìƒí–ˆìŠµë‹ˆë‹¤. API í‚¤ë¥¼ í™•ì¸í•´ì£¼ì„¸ìš”.")
    # 10 ì‹¬í”Œí•œ í‘¸í„°
    st.markdown("---")
    st.caption("Made with â¤ï¸ by AI Copywriter")
    
if __name__ == "__main__":
    main()
```
